### 1. Constante (O(1))

A complexidade constante significa que o tempo de execução não depende do tamanho da entrada.

```python

def constante(n):
    return n * 2  # Operação constante
constante(5)
```


### 2. Linear (O(n))

A complexidade linear significa que o tempo de execução cresce proporcionalmente ao tamanho da entrada.

```python

def linear(n):
    for i in range(n):
        print(i)  # Operação linear

linear(5)
```

### 3. Logarítmica (O(log n))

A complexidade logarítmica é frequentemente associada a algoritmos de divisão e conquista, como a busca binária.

```python
def logaritmica(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

logaritmica([1,2,3,4,5],3)
```

### 4. Logarítmica Linear (O(n log n))

A complexidade logarítmica linear é comum em algoritmos de ordenação eficientes como o Merge Sort e o Quick Sort.

```python
def n_log_n(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = n_log_n(arr[:mid])
    right = n_log_n(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

n_log_n([1,10,3,4,20,6,7,8,9,2])
```

### 5. Quadrática (O(n^2))

A complexidade quadrática é comum em algoritmos que usam dois loops aninhados.

```python

def quadratica(arr):
    for i in range(len(arr)):
        for j in range(len(arr)):
            print(arr[i], arr[j])  # Operação quadrática

quadratica([5,5])
```

### 6. Exponencial (O(2^n))

A complexidade exponencial é comum em problemas que têm múltiplas escolhas em cada etapa, como na solução de problemas combinatórios.

```python

def exponencial(n):
    if n <= 1:
        return 1
    else:
        return exponencial(n - 1) + exponencial(n - 1)

exponencial(5)
```

### 7. Fatorial (O(n!))

A complexidade fatorial é comum em algoritmos que geram todas as permutações possíveis

```python

def fatorial(n):
    if n == 0:
        return 1
    else:
        return n * fatorial(n - 1)

fatorial(5)
```

### 8. Polinomial (O(n^k))

A complexidade polinomial é uma generalização para qualquer função polinomial. Um exemplo comum é um algoritmo de força bruta para resolver um problema.

```python

def polinomial(arr):
    n = len(arr)
    for i in range(n):           # O(n)
        for j in range(n):       # O(n)
            for k in range(n):   # O(n)
                print(arr[i], arr[j], arr[k])  # Operação polinomial O(n^3)

polinomial([5,5,5])
```

```python
#Desempenho
import time
import matplotlib.pyplot as plt
import numpy as np

# Função para medir o tempo de execução
def medir_tempo(algoritmo, *args):
    inicio = time.time()
    algoritmo(*args)
    fim = time.time()
    return fim - inicio

# Função para plotar o gráfico de desempenho
def plotar_desempenho():
    tamanhos = [10, 100, 1000, 10000]
    tempos_constante = []
    tempos_linear = []
    tempos_logaritmica = []
    tempos_n_log_n = []
    tempos_quadratica = []

    for n in tamanhos:
        # Prepara dados de entrada para cada algoritmo
        arr = list(range(n))
        target = n // 2

        tempos_constante.append(medir_tempo(constante, n))
        tempos_linear.append(medir_tempo(linear, n))
        tempos_logaritmica.append(medir_tempo(logaritmica, arr, target))
        tempos_n_log_n.append(medir_tempo(n_log_n, arr))
        tempos_quadratica.append(medir_tempo(quadratica, arr[:int(np.sqrt(n))]))  # Limitando o tamanho para quadratica

    # Plotando os resultados
    plt.figure(figsize=(12, 8))

    plt.plot(tamanhos, tempos_constante, label="Constante O(1)")
    plt.plot(tamanhos, tempos_linear, label="Linear O(n)")
    plt.plot(tamanhos, tempos_logaritmica, label="Logarítmica O(log n)")
    plt.plot(tamanhos, tempos_n_log_n, label="Linearítmica O(n log n)")
    plt.plot(tamanhos, tempos_quadratica, label="Quadrática O(n^2)")

    plt.xscale('log')
    plt.yscale('log')

    plt.xlabel('Tamanho de n (log scale)')
    plt.ylabel('Tempo de Execução (s) (log scale)')
    plt.title('Desempenho dos Algoritmos por Complexidade')
    plt.legend()
    plt.grid(True, which="both", ls="--")

    plt.show()

# Executar a função para plotar o gráfico
plotar_desempenho()
```
